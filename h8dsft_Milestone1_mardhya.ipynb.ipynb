#### HOW

```python
from scipy import stats
std = df['Total_Sales'].std()
N = len(df)
low, up = stats.norm.interval(0.95,loc=df['Total_Sales'].mean(),scale=std/np.sqrt(N))
print('Lower Limit:',low)
print('Upper Limit:',up)
```

Conducting a Chi-Squared Hypothesis Test

H0: Tidak ada hubungan diantara Jumlah Produk dan Harga_Produk Yang Mempengaruhi total penjualan.

H1: Ada hubungan diantara Jumlah Produk dan Harga_Produk yang mempengaruhi total penjualan.

```python
contingency_table = pd.crosstab(df['Price_PerUnit'],df['Units_Sold'])
res = stats.chi2_contingency(contingency_table)
print("P-value:",res.pvalue)
```

Critical value: 0.05

p-val = 2.7402880705029854e-119 < critical value

H0 Ditolak

Conclusion:
Ada hubungan diantara Jumlah Produk dan Harga_Produk yang mempengaruhi total penjualan.

```python
q1 = df['Units_Sold'].quantile(0.25)
q3 = df['Units_Sold'].quantile(0.75)
iqr = q3-q1

up_bound = q3 + 1.5*iqr
low_bound = q1 - 1.5*iqr

print(f'Upper Boundary: {up_bound:.2f}' )
print(f'Lower Boundary: {low_bound:.2f}')

# filter data outlier
outlier = df['Units_Sold'][(df['Units_Sold']<low_bound) | (df['Units_Sold']>up_bound)]

# filter data non-outlier
no_outlier = df['Units_Sold'][(df['Units_Sold']>=low_bound) & (df['Units_Sold']<=up_bound)]

print(f'Percentage of outliers: {len(outlier)*100/len(df["Units_Sold"]):.2f}%%')
```

```python
plt.figure(figsize=(6, 4))
plt.boxplot(df['Units_Sold'], vert=False, patch_artist=True)

plt.annotate(f'Q1 = {q1}', xy=(q1, 1.06), xytext=(q1, 1.2), arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.annotate(f'Q3 = {q3}', xy=(q3, 1.06), xytext=(q3, 1.2), arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.annotate(f'IQR = {iqr}', xy=((q1+q3)/2, 0.85), ha='center')

plt.axvline(q1, linestyle='--', color='gray', label='Q1')
plt.axvline(q3, linestyle='--', color='gray', label='Q3')
plt.axvspan(q1, q3, alpha=0.2, color='gray', label='IQR')

plt.xlabel('Values')
plt.title('Boxplot of Units Sold')

plt.show()
```
